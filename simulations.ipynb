# Original Model

```python
importlib.reload(dialog_logic)

from dialog_logic import *

p = 1.0
q = 1.0
transition = np.eye(2)
transition[0, 0] = p
transition[0, 1] = 1 - p
transition[1, 0] = 1 - q
transition[1, 1] = q
# print_transition_matrix(transition)

alice = [1, 0, 0]
bob   = [0, 1, 0]

alice_results = []
bob_results = []
# convert to floats
alice = np.array(alice).astype(float)
bob = np.array(bob).astype(float)

alice_results.append(alice)
bob_results.append(bob)
# print("Initial joint preference")
# print(make_joint_preference(alice, bob))

# Communicate
iterations = 5
for i in range(iterations):
    # make 2x3 matrix from alice and bob
    dialog_matrix = [alice, bob]

    result = np.dot(transition, dialog_matrix)
    print("\nResult")
    print(result)

    # marginalize result to get Alice's opinion
    alice = result[0]
    # marginalize result to get Bob's opinion
    bob = result[1]

    # print("\nBob's opinion")
    # print(bob, "->", bob_result)
    alice_results.append(alice)
    bob_results.append(bob)
    # return alice_result, bob_result
plot_one(alice_results, f"original-model-alice", "Alice")
plot_one(bob_results, f"original-model-bob", "Bob")
```
